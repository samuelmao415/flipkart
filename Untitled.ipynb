```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
Clothing["percentage"]=(toy["retail_price"]-toy["discounted_price"])/(toy["retail_price"])
```

```python
Clothing=Clothing.reset_index()
```

```python
Clothing.percentage
```

```python

cloth_plot=Clothing.groupby("brand").agg({'percentage': 'mean'}).sort_values(by="percentage",ascending=False)
cloth_plot.plot.bar(figsize=(25,8),color="#c44e52",legend=False)


plt.xlabel("Brand")
plt.ylabel("Least discounted brand")
```
